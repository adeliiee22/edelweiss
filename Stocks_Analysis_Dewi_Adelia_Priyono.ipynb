# Stocks Metrics Analysis 


1.   **Average ratio of price difference to closing price**

      Calculating the average proportion of the difference between the high and low prices compared to the closing price to **captures a specific aspect of price movement within a day**.

2.   **Volality**

      Volatility is a **measure of risk and uncertainty in the stock market**. High volatility implies that the price of a stock can change significantly in a short period, indicating greater potential for both gains and losses. On the other hand, low volatility suggests that the stock price is relatively stable and does not experience significant price swings.

```python
import numpy as np
!pip install pandas==1.3.5 a
import pandas as pd
!pip install xlrd==1.2.0
import yfinance as yf

# package to download data, from pandas 
import pandas_datareader.data as web

# packages to work with time and dates

import datetime as dt
import matplotlib.dates as mdates

# packages for visulization

import matplotlib.pyplot as plt
plt.style.use('ggplot')               
```

# Unilever Stock Analysis

### Average ratio of price difference to closing price

```python
start = '2021-01-01'   # Start date to download the data
end = '2023-01-01'     # End Date to download the data

UNVR_JK = yf.download('UNVR.JK', start=start, end=end)

# Calculate the difference between high and low prices
UNVR_JK['Price_Difference'] = UNVR_JK['High'] - UNVR_JK['Low']

# Calculate the ratio of price difference to the closing price
UNVR_JK['Ratio'] = UNVR_JK['Price_Difference'] / UNVR_JK['Close']

# Calculate the average ratio
average_ratio = np.mean(UNVR_JK['Ratio'])

print("Average ratio of price difference to closing price: %.2f" % average_ratio)
```

If the average ratio of price difference to closing price is 0.03, it means that, on average, **the difference between the high and low prices of the stock is approximately 3% of the closing price**.

<br>

A **higher average ratio suggests larger price swings or volatility within each trading day**, while a lower average ratio indicates relatively smaller price movements.

### Volatility

```python
stock = yf.Ticker('UNVR.JK')
start_date = "2021-01-01"
end_date = "2022-12-31"
data = stock.history(start=start_date, end=end_date)

data["Daily_Return"] = data["Close"].pct_change()

#calculate tge daily volatility
daily_volatility = data["Daily_Return"].std()

# Calculate the average volatility

print("Daily Volatility: ", daily_volatility.round(2))
```

A volatility of 0.02 implies that the price of the stock typically experiences smaller fluctuations compared to a stock with higher volatility. The stock's price tends to **remain relatively stable** over the given period, with less rapid and significant price changes.

### Volatility Graph

```python
fig, ax = plt.subplots(figsize=(10, 8))
plt.plot(data["Daily_Return"], color="blue")  
plt.title("Daily Returns of Stock")
plt.ylabel("Daily Return")
plt.xlabel("Date")
plt.show()
```
